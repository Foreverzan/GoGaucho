# UCSB Instructor Grade Given Analysis

Our team is developing a new UCSB app called "Gogaucho" which have many features, like map, class schedule, Instructors,etc. This analysis is construct a database which contains all the UCSB instructor grade given from the past quarter that give students idea how difficult of the course is to help them choose instructor.

```python
import numpy as np 
import pandas as pd 
import csv
```

#### Import dataset "grades.xlsx"

```python
df = pd.concat(pd.read_excel('grades.xlsx', sheet_name=None), ignore_index=True)
df.head(5)
```

#### Cleaning & Group

```python
df.drop(['Course Level','Course'],axis=1).groupby(['Instructor']).count()
```

```python
df2 = df.drop(['Course Level','Course'],axis=1).groupby(['Instructor','Grade Given']).count()
df2.head(10)
```

```python
df3 = df2.reset_index(inplace=False)
df3.head(5)
```

```python
df4 = df3.pivot(index='Grade Given', columns='Instructor', values='Sum of Student Count').fillna(0)
df4.head(5)
```

```python
df5 = df4.to_dict()
#df5
```
